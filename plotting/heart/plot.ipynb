```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# import numpy as np

# Load dataset
df = pd.read_csv("heart.csv")
# sns.set(style="whitegrid")

# a. Histogram  Objective: Visualize distribution of age

plt.hist(df['age'])
plt.title("Histogram: Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# b. Dot Plot  Objective: Show relationship between Age and Cholesterol
plt.figure(figsize=(8, 5))
plt.plot(df['age'], df['chol'], 'o')
plt.title("Dot Plot: Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.grid(True)
plt.show()

# c. Bar Plot  Objective: Compare average cholesterol across chest pain types
plt.figure(figsize=(8, 5))
sns.barplot(x='cp', y='chol', data=df, palette='magma')
plt.title("Bar Plot: Chest Pain Type vs Average Cholesterol")
plt.xlabel("Chest Pain Type")
plt.ylabel("Average Cholesterol")
plt.show()

# d. Line Chart  Objective: Observe trend of resting blood pressure over sorted age
df_sorted = df.sort_values('age')
plt.figure(figsize=(8, 5))
plt.plot(df_sorted['age'], df_sorted['trestbps'], marker='o', linestyle='-', color='green')
plt.title("Line Chart: Resting Blood Pressure vs Age")
plt.xlabel("Age")
plt.ylabel("Resting Blood Pressure")
plt.grid(True)
plt.show()

# e. Combined Plot  Objective: Visualize distribution and outliers in 'thalach'
fig = plt.figure(figsize=(12, 8))
gs = fig.add_gridspec(3, 3)
ax_box = fig.add_subplot(gs[0, 1:])
ax_hist = fig.add_subplot(gs[1:, :-1])
ax_scatter = fig.add_subplot(gs[1:, 1:])

# Boxplot
sns.boxplot(y=df['thalach'], ax=ax_box, color='lightblue')
ax_box.set_title("Box Plot: Max Heart Rate Achieved")

# Histogram
sns.histplot(df['thalach'], kde=True, ax=ax_hist, color='skyblue')
ax_hist.set_xlabel("Max Heart Rate")
ax_hist.set_ylabel("Frequency")

# Scatter Plot
sns.scatterplot(x='age', y='thalach', data=df, ax=ax_scatter, color='red')
ax_scatter.set_title("Scatter Plot: Age vs Max Heart Rate")

plt.tight_layout()
plt.show()
```
