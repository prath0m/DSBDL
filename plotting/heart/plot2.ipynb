```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("heart.csv")
sns.set(style="whitegrid")

# a. Histogram – Objective: Understand distribution of numerical features like 'age'
plt.figure(figsize=(8, 5))
plt.hist(df['age'], bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram: Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# b. Pie Chart – Objective: Show percentage of patients with and without heart disease
plt.figure(figsize=(6, 6))
target_counts = df['target'].value_counts()
labels = ['No Heart Disease', 'Heart Disease']
colors = ['lightcoral', 'lightgreen']
plt.pie(target_counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title("Pie Chart: Heart Disease Distribution")
plt.axis('equal')
plt.show()

# c. Box Plot – Objective: Detect outliers in cholesterol levels
plt.figure(figsize=(8, 5))
sns.boxplot(y='chol', data=df, color='lightblue')
plt.title("Box Plot: Cholesterol Levels")
plt.ylabel("Cholesterol")
plt.show()

# d. Scatter Plot – Objective: Relationship between age and maximum heart rate
plt.figure(figsize=(8, 5))
sns.scatterplot(x='age', y='thalach', hue='target', data=df, palette='coolwarm')
plt.title("Scatter Plot: Age vs Max Heart Rate by Target")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate (thalach)")
plt.legend(title='Heart Disease')
plt.show()

# e. Boxplots on a Scatterplot – Objective: Visualize distribution and relationship together
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='thalach', data=df, width=0.3, palette='Set2')
sns.stripplot(x='target', y='thalach', data=df, jitter=0.2, color='black', alpha=0.5)
plt.title("Combined Boxplot and Scatter: Max Heart Rate vs Heart Disease")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Max Heart Rate (thalach)")
plt.show()
```
